```python
from graphviz import Digraph

dot = Digraph()
dot.attr(ranksep='0.75', nodesep='0.5', labelloc='t')

dot.node('a',  '', xlabel='s', shape='circle')
dot.node('b1', '', xlabel='a', shape='circle', style='filled', color='black')
dot.node('b2', '', shape='circle', style='filled', color='black')
dot.node('b3', '', shape='circle', style='filled', color='black')
dot.node('c1', '', shape='circle')
dot.node('c2', '', shape='circle')
dot.node('c3', '', shape='circle')
dot.node('c4', '', shape='circle')
dot.node('c5', '', xlabel='s\'', shape='circle')
dot.node('c6', '', shape='circle')

dot.edge('a', 'b1')
dot.edge('a', 'b2', xlabel='<&pi;>')
dot.edge('a', 'b3')
dot.edge('b1', 'c1')
dot.edge('b1', 'c2')
dot.edge('b2', 'c3')
dot.edge('b2', 'c4')
dot.edge('b3', 'c5', label='p')
dot.edge('b3', 'c6', label='r')

dot
```
